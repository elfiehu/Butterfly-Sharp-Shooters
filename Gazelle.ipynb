```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
year_2016 = "Resources/2016data.csv"
year_2017 = "Resources/2017data.csv"

df_2016 = pd.read_csv(year_2016)
df_2017 = pd.read_csv(year_2017)
```

```python
df_2016.head()
```

```python
df_2017.head()
```

```python
category_group_2016 = df_2016.groupby('Category Description') 
sum_category_2016 = category_group_2016['quantity'].sum()
piechart_data2016 = sum_category_2016.drop(sum_category_2016.index[[0,8]])
piechart_data2016
```

```python
top_2016_chart = piechart_data2016.plot(kind='pie',figsize=(14,14), autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
plt.ylabel("")
plt.title('Top Categories in 2016')
plt.savefig('top_categories_2016.png')
plt.show()
```

```python
category_group_2017 = df_2017.groupby('Category Description') 
sum_category_2017 = category_group_2017['quantity'].sum()
piechart_data2017 = sum_category_2017.drop(sum_category_2016.index[[0,1]])
piechart_data2017
```

```python
top_2017_chart = piechart_data2017.plot(kind='pie',figsize=(14,14), autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
plt.ylabel("")
plt.title('Top Categories in 2017')
plt.savefig('top_categories_2017.png')
plt.show()
```
